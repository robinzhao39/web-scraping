```python
import pandas as pd
base_url='http://linserv1.cims.nyu.edu:10000/films?_page='
i=1
final_list=[]
while True:
    url=base_url+str(i)
    r=requests.get(url)
    if r.json()==[]:
        break
    final_list+=r.json()
    i+=1
df=pd.DataFrame(final_list)
```

```python
df
```

## 3

```python
df['rt_score'].astype('float')
```

```python
df['rt_score']=df['rt_score'].astype('float')
s1=df['rt_score'].groupby(df['director']).mean()
```

```python
s2=df.groupby(df['director'])['title'].count()
```

```python
report=pd.concat([s1,s2],axis=1)
report=report.rename({'rt_score':'avg_rt_score','title':'count'},axis=1)
report
```
