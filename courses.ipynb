```python
from bs4 import BeautifulSoup
import requests
import re
```

## 1

```python
with open('schedule_link.html', 'r') as f:
    contents = f.read()
    dom = BeautifulSoup(contents)
    rows=dom.select('.schedule-listing .row')
    dataframe=[]
    for row in rows:
        new_dic={}
        string=row.select('span')[0].text.strip()
        new_dic['Number_section']=re.match('\w*-\w\w.\d{4}-\u200b\d{3}',string)[0].replace('\u200b','')
        raw_p=row.select('span')[2].text.strip().split('\n')
        processed_p=[]
        for x in raw_p:
            x=x.strip()
            if x!='' and x!='Office Hours':
                processed_p.append(x)
        new_dic['Instructor']=processed_p
        string=row.select('span')[1].text.strip().replace('\n','')
        new_dic['Name']=" ".join(string.split())
        new_dic['Time']=row.select('span')[3].text.strip()
        dataframe.append(new_dic)
    
```

```python
import pandas as pd
df=pd.DataFrame(dataframe)
df['Section']=df['Number_section'].str.extract('\d{4}-(\d{3})')
df['Number']=df['Number_section'].str.extract('(\w{4}-\w\w.\d{4})-\d{3}')
df.info()
```

## 2

```python
with open('course_catalog.html', 'r') as f:
    contents = f.read()
    dom = BeautifulSoup(contents)
    dataframe=[]
    rows=dom.select('.courses-listing .col-sm-12')
    for row in rows:
        new_dic={}
        course=row.select('p')[0].text.strip()
        new_dic['Number']=re.match('\w{4,5}-\w\w.\d{4}',course)[0]
        point=row.select('p')[1].text.strip()
        new_dic['Points']=re.match('.*Points',point)[0].replace('Points','')
        pre=row.select('p')[2].text.strip()
        new_dic['Prereqs']=pre[15:]
        dataframe.append(new_dic) 
```

```python
df1=pd.DataFrame(dataframe)
df1
```
